```python
def compile(bytecodes, idx, num_indent, elem_no, check_flag):
    pycode = []
    indent = "    "
    prev_cur = 0
    i = 0
    cursor = 0
    code = bytecodes[idx]
    check_flag = False
    while cursor < len(code):
        if code[cursor] == 0:#push
            if elem_no ==(10000-1) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-1:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 1
            cursor += 1
            pycode.append(indent*num_indent+f"cur = stack_push(stack, {code[cursor]}, cur)\n")
            cursor += 1
        elif code[cursor] == 3: #if-then
            if elem_no < 1 or check_flag:
                pycode.append(indent*num_indent+"if cur == 0:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 1
            pycode.append(indent*num_indent+f"temp, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"if temp != 0:\n")
            num_indent += 1
            cursor += 1
            func_address = code[cursor]-71
            numba_code, elem_no = compile(bytecodes,func_address, num_indent, elem_no, check_flag)
            pycode.extend(numba_code)            
            num_indent -= 1
            cursor += 1
        elif code[cursor] == 4:#if-else-then
            if elem_no < 1 or check_flag:
                pycode.append(indent*num_indent+"if cur == 0:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            pycode.append(indent*num_indent+f"temp, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"if temp != 0:\n")
            num_indent += 1
            cursor += 1
            func_address = code[cursor]-71
            numba_code, elem_no = compile(bytecodes,func_address, num_indent, elem_no, check_flag)
            pycode.extend(numba_code) 
            num_indent -= 1
            pycode.append(indent*num_indent+f"else:\n")
            cursor += 1
            num_indent += 1
            func_address = code[cursor]-71
            numba_code, elem_no = compile(bytecodes,func_address, num_indent, elem_no, check_flag)
            pycode.extend(numba_code)            
            num_indent -= 1
            cursor += 1
        elif code[cursor] == 6: #do-loop?
            check_flag = True
            if elem_no < 2 or check_flag:
                check_flag = True
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"for i in range(temp1,temp2):\n")
            num_indent += 1
            cursor += 1
            func_address = code[cursor]-71
            numba_code, elem_no = compile(bytecodes,func_address, num_indent, elem_no, check_flag)
            pycode.extend(numba_code)            
            num_indent -= 1
            cursor += 1
        elif code[cursor] == 7: #do-loop
            check_flag = True
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"for i in range(temp1,temp2):\n")
            num_indent += 1
            cursor += 1
            func_address = code[cursor]-71
            numba_code, elem_no = compile(bytecodes,func_address, num_indent, elem_no, check_flag)
            pycode.extend(numba_code)
            num_indent -= 1
            cursor += 1
        elif code[cursor] == 37:#dup
            if elem_no < 1 or check_flag:
                pycode.append(indent*num_indent+"if cur < 1:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 1
            pycode.append(indent*num_indent+f"temp, cur = stack_pop(stack, cur)\n")
            cursor += 1
            if elem_no ==(10000-2) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp, cur)\n")
            elem_no += 2
                
        elif code[cursor] == 38:#pop and drop
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 1
            pycode.append(indent*num_indent+f"temp, cur = stack_pop(stack, cur)\n")
            cursor += 1
        elif code[cursor] == 39:# swap
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            cursor += 1
            if elem_no ==(10000-2) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 2
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp2, cur)\n")
        elif code[cursor] == 40:# over
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            cursor += 1
            if elem_no ==(10000-3) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-3:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 3
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp2, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp2, cur)\n")
        elif code[cursor] == 41:# rot
            if elem_no < 3 or check_flag:
                pycode.append(indent*num_indent+"if cur < 3:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 3
            pycode.append(indent*num_indent+f"temp1, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp3, cur = stack_pop(stack, cur)\n")
            cursor += 1
            if elem_no ==(10000-3) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-3:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 3
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp2, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp3, cur)\n")
        elif code[cursor] == 42:# nip
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1, cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            cursor += 1
            if elem_no ==(10000-1) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-1:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 1
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1, cur)\n")
        elif code[cursor] == 43:#tuck
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1,cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            if elem_no ==(10000-3) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-3:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 1
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp2, cur)\n")
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1, cur)\n")
            cursor += 1
        elif code[cursor] == 44:#add
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1,cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            if elem_no ==(10000-1) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-1:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 1
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp1+temp2, cur)\n")
            cursor += 1
        elif code[cursor] == 45:#subtract
            if elem_no < 2 or check_flag:
                pycode.append(indent*num_indent+"if cur < 2:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no -= 2
            pycode.append(indent*num_indent+f"temp1,cur = stack_pop(stack, cur)\n")
            pycode.append(indent*num_indent+f"temp2, cur = stack_pop(stack, cur)\n")
            if elem_no ==(10000-1) or check_flag:
                pycode.append(indent*num_indent+"if cur == len(stack)-1:\n")
                pycode.append(indent*(num_indent+1)+"raise ValueError('Index out of range')\n")
            elem_no += 1
            pycode.append(indent*num_indent+f"cur = stack_push(stack, temp2-temp1, cur)\n")
            cursor += 1
    return pycode, elem_no
```

```python
init = ["@jit(nopython=True, nogil=True)\n", "def func(stack):\n", "    cur = np.uint64(0)\n"]
pycode, elem_no = compile(ak.to_list(bytecodes),0 ,1, 0, False)
#pycode.append("    print(stack, cur)\n")
#pycode.append("func()")
init.extend(pycode)
aa = "".join(init)
```

```python
@jit(nopython=True, nogil=True)
def stack_push(stack,elem, cur):
#     if cur == len(stack)-1:
#         raise ValueError("Index out of range")
    stack[cur] = elem
    cur += 1
    return cur

@jit(nopython=True, nogil=True)
def stack_pop(stack, cur):
#     if cur == 0:
#         raise ValueError("Index out of range")
    cur -= 1
    return stack[cur], cur
```
